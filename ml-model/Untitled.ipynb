```python
# Avoid divide by zero
df['Balls'] = df['Balls'].replace(0, 1)

# Strike rate
df['strike_rate'] = (df['Runs'] / df['Balls']) * 100

# Consistency
df['Inn'] = df['Inn'].replace(0, 1)

df['consistency'] = df['Runs'] / df['Inn']
```

```python
df['batsman_score'] = (
    df['Avg'] * 0.40 +
    df['strike_rate'] * 0.30 +
    df['Runs'] * 0.20 +
    df['consistency'] * 0.10
)
```

```python
df['bowler_score'] = (
    df['Wkt'] * 0.50 +
    (1 / (df['Avg_2'] + 1)) * 100 * 0.30 +
    df['Mdn'] * 0.20
)
```

```python
bowling_features = [
    'Wkt',
    'Avg_2',
    'Mdn',
    'economy',
    'Mat'
]

recent_df[bowling_features] = scaler.fit_transform(
    recent_df[bowling_features]
)
```
